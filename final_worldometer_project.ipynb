```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/','html.parser')
response = requests.get(url)
#print(response)
data = response.text
#print(data)
soup = BeautifulSoup(data,'lxml')
#print(soup)
tags = soup.table
tag = tags.find_all('tr')
#print(tag)
a = []
for i in tag:
   s = i.get_text()
   a.append(s)
#print(a)
final = a[9:19]
final_data = []
for i in final:
  f = i.split('\n')[1:-1]
  final_data.append(f)
#print(final_data)
final_data = [['SN', 'Country name', 'TotalCases', 'NewCases', 'TotalDeaths', 'NewDeaths', 'TotalRecovered', 'NewRecovered', 'ActiveCases', 'Serious,Critical', 'TotalCases/1M pop', 'Deaths/1M pop', 'TotalTests', 'Tests/1M pop', 'Population']] + final_data

import csv
n = len(final_data)
f = open('worldometer.csv','w')
w = csv.writer(f)
for i in range(n):
  w.writerow(final_data[i])
f.close()

import pandas as pd
l = ['SN', 'Country name', 'TotalCases', 'NewCases', 'TotalDeaths', 'NewDeaths', 'TotalRecovered', 'NewRecovered', 'ActiveCases', 'Serious,Critical', 'TotalCases/1M pop', 'Deaths/1M pop', 'TotalTests', 'Tests/1M pop', 'Population']
df = pd.read_csv('worldometer.csv', index_col="SN",usecols = l)
final = df.fillna(0)
final


```

```python
import plotly.express as px
fig = px.bar(final, x='Country name', y='TotalCases',
             hover_data=['Country name', 'TotalCases'], color='Country name',
             labels={'TC':'TotalCases'}, height=400)
fig.show()
```
