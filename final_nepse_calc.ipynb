```python
from bs4 import BeautifulSoup
import requests
d = []
n = int(input("Enter n less than 7 = "))
for i in range(1,n):
  url = BeautifulSoup(f"http://www.nepalstock.com/main/calculation/index/{i}/", 'html.parser')
  response = requests.get(url)
  data = response.text
  soup = BeautifulSoup(data,'xml')
  tags = soup.table
  tag = tags.find_all('tr')
  a = []
  for i in tag:
    s = i.get_text()
    a.append(s)
  final_data = a[2:-1]
  
  for i in final_data:
    f = i.split('\n')[1:-1]
    d.append(f)

d = [['S.N.','Stock Symbol','Closing Price Average','Total Traded Amount','Total Traded Shares','Weighted Average','Closing Price','Closing Date']] + d
import csv
num = len(d)
file = open('finalnepsecalc.csv','w')
writer = csv.writer(file)
for j in range(num):
  writer.writerow(d[j])
file.close()

import pandas as pd
df = pd.read_csv('finalnepsecalc.csv', index_col='S.N.')
df
```
